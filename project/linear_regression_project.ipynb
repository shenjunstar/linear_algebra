```python
# TODO 返回矩阵的行数和列数
def shape(M):
    _set = set()
    for i in range(len(M)):
        _set.add(len(M[i]))
    if len(_set) > 1 or len(_set) <= 0 :
        raise ValueError('Matrix has wrong length of row')
    return (len(M),_set.pop())
```

## 1.5 计算矩阵乘法 AB

```python
# TODO 计算矩阵乘法 AB，如果无法相乘则raise ValueError
def matxMultiply(A, B):
    shape_a = shape(A)
    shape_b = shape(B)
    if shape_a[1] != shape_b[0]:
        raise ValueError('Matrix a cannot multiply matrix b')
    column_b_arr = []
    new_matrix = []
    for i in range(shape_a[0]):
        row_a = A[i]
        new_row = []
        if i == 0:
            for j in range(shape_b[1]):
                column_b = [x[j] for x in B]
                result = sum([x*y for x,y in zip(row_a,column_b)])
                new_row.append(result)
                column_b_arr.append(column_b)
        else :
            for column_b in column_b_arr:
                result = sum([x*y for x,y in zip(row_a,column_b)])
                new_row.append(result)
        new_matrix.append(new_row)
    return new_matrix
```

## 2.1 构造增广矩阵

$ A = \begin{bmatrix}
    a_{11}    & a_{12} & ... & a_{1n}\\
    a_{21}    & a_{22} & ... & a_{2n}\\
    a_{31}    & a_{22} & ... & a_{3n}\\
    ...    & ... & ... & ...\\
    a_{n1}    & a_{n2} & ... & a_{nn}\\
\end{bmatrix} , b = \begin{bmatrix}
    b_{1}  \\
    b_{2}  \\
    b_{3}  \\
    ...    \\
    b_{n}  \\
\end{bmatrix}$

返回 $ Ab = \begin{bmatrix}
    a_{11}    & a_{12} & ... & a_{1n} & b_{1}\\
    a_{21}    & a_{22} & ... & a_{2n} & b_{2}\\
    a_{31}    & a_{22} & ... & a_{3n} & b_{3}\\
    ...    & ... & ... & ...& ...\\
    a_{n1}    & a_{n2} & ... & a_{nn} & b_{n} \end{bmatrix}$

```python
# TODO 构造增广矩阵，假设A，b行数相同
def augmentMatrix(A, b):
    _shape = shape(A)
    new_matrix = []
    for i in range(_shape[0]):
        rowA = A[i].copy()
        rowA.extend(b[i])
        new_matrix.append(rowA)
    return new_matrix
```

## 2.3  Gaussian Jordan 消元法求解 Ax = b

### 2.3.1 算法

步骤1 检查A，b是否行数相同

步骤2 构造增广矩阵Ab

步骤3 逐列转换Ab为化简行阶梯形矩阵 [中文维基链接](https://zh.wikipedia.org/wiki/%E9%98%B6%E6%A2%AF%E5%BD%A2%E7%9F%A9%E9%98%B5#.E5.8C.96.E7.AE.80.E5.90.8E.E7.9A.84-.7Bzh-hans:.E8.A1.8C.3B_zh-hant:.E5.88.97.3B.7D-.E9.98.B6.E6.A2.AF.E5.BD.A2.E7.9F.A9.E9.98.B5)
    
    对于Ab的每一列（最后一列除外）
        当前列为列c
        寻找列c中 对角线以及对角线以下所有元素（行 c~N）的绝对值的最大值
        如果绝对值最大值为0
            那么A为奇异矩阵，返回None (你可以在选做问题2.4中证明为什么这里A一定是奇异矩阵)
        否则
            使用第一个行变换，将绝对值最大值所在行交换到对角线元素所在行（行c） 
            使用第二个行变换，将列c的对角线元素缩放为1
            多次使用第三个行变换，将列c的其他元素消为0
            
步骤4 返回Ab的最后一列

**注：** 我们并没有按照常规方法先把矩阵转化为行阶梯形矩阵，再转换为化简行阶梯形矩阵，而是一步到位。如果你熟悉常规方法的话，可以思考一下两者的等价性。

### 2.3.3 实现 Gaussian Jordan 消元法

```python
# TODO 实现 Gaussain Jordan 方法求解 Ax = b
import copy

""" Gaussian Jordan 方法求解 Ax = b.
    参数
        A: 方阵 
        b: 列向量
        decPts: 四舍五入位数，默认为4
        epsilon: 判读是否为0的阈值，默认 1.0e-16
        
    返回列向量 x 使得 Ax = b 
    返回None，如果 A，b 高度不同
    返回None，如果 A 为奇异矩阵
"""
def gj_Solve(A, b, decPts=4, epsilon=1.0e-16):
    # 检查A,b行数是否相同
    shapeA = shape(A)
    shapeB = shape(b)
    if shapeA[0] != shapeB[0]:
        raise ValueError('Row number of A must equals row number of b')
    #构造增广矩阵
    argM = augmentMatrix(A,b)
    copyM = copy.deepcopy(argM)
    #遍历每一列，除了最后一列
    for i in range(shapeA[0]):
        _column = [x[i] for x in copyM]
        sliceColumn = _column[i::]
        _maxOfColumn = max([abs(x) for x in sliceColumn])
        #如果是奇异矩阵则无解
        if _maxOfColumn < epsilon:
            return None
        if abs(copyM[i][i]) < epsilon:
            #交换行
            _firstNoneZeroIndex = first_none_zero_index(_column,i,epsilon)
            swapRows(copyM,i,_firstNoneZeroIndex)
        #对角线元素置为1
        scaleRow(copyM,i,1./copyM[i][i])
        for j in range(shapeA[0]):
            #r1 <--- r1 + r2*scale
            if j != i:
                if abs(copyM[j][i]) > epsilon:
                    addScaledRow(copyM, j, i, -(copyM[j][i]))
    calc_column_array = column_gains(copyM)
    resultM = calc_column_array[len(calc_column_array)-1]
    resultM = [[resultM[j] for i in range(1)] for j in range(len(resultM))]
    matxRound(resultM,decPts)
    return resultM

def column_gains(M):
    columns = []
    for i in range(len(M[0])):
        columns.append([x[i] for x in M])
    return columns

def first_none_zero_index(arr,limitIndex,epsilon=1.0e-16):
    for i in range(len(arr)):
        if abs(arr[i]) > epsilon and i > limitIndex:
            return i
    return None
```

### 3.2.2 计算平均平方误差 (MSE)

我们要编程计算所选直线的平均平方误差(MSE), 即数据集中每个点到直线的Y方向距离的平方的平均数，表达式如下：
$$
MSE = \frac{1}{n}\sum_{i=1}^{n}{(y_i - mx_i - b)^2}
$$

```python
# TODO 实现以下函数并输出所选直线的MSE
def calculateMSE2D(X,Y,m,b):
    lenX = len(X)
    lenY = len(Y)
    if lenX != lenY:
        raise ValueError("Length of X must equals length of Y")
    mse = sum([(y-m*x-b)**2 for x,y in zip(X,Y)])/lenX
    return mse

# TODO 检查这里的结果, 如果你上面猜测的直线准确, 这里的输出会在1.5以内
print(calculateMSE2D(X,Y,m1,b1))
```

## 3.3 (选做) 找到参数 $m, b$ 使得平方平均误差最小

**这一部分需要简单的微积分知识(  $ (x^2)' = 2x $ )。因为这是一个线性代数项目，所以设为选做。**

刚刚我们手动调节参数，尝试找到最小的平方平均误差。下面我们要精确得求解 $m, b$ 使得平方平均误差最小。

定义目标函数 $E$ 为
$$
E = \frac{1}{2}\sum_{i=1}^{n}{(y_i - mx_i - b)^2}
$$

因为 $E = \frac{n}{2}MSE$, 所以 $E$ 取到最小值时，$MSE$ 也取到最小值。要找到 $E$ 的最小值，即要找到 $m, b$ 使得 $E$ 相对于 $m$, $E$ 相对于 $b$ 的偏导数等于0. 

因此我们要解下面的方程组。

$$
\begin{cases}
\displaystyle
\frac{\partial E}{\partial m} =0 \\
\\
\displaystyle
\frac{\partial E}{\partial b} =0 \\
\end{cases}
$$

$$
\frac{\partial E}{\partial m} = \sum_{i=1}^{n}{-x_i(y_i - mx_i - b)}
$$

$$
\frac{\partial E}{\partial b} = \sum_{i=1}^{n}{-(y_i - mx_i - b)}
$$

   \sum_{i=1}^{n}{x_iy_i} \\
   \sum_{i=1}^{n}{y_i}\\
\end{bmatrix}
$$
那么：
$$
X^TXh - X^TY = \begin{bmatrix}
   \sum_{i=1}^{n}{mx_i^2+bx_i-x_iy_i} \\
   \sum_{i=1}^{n}{mx_i+b-y_i}\\
\end{bmatrix} = \begin{bmatrix}
   \sum_{i=1}^{n}{-x_i(y_i-mx_i-b)} \\
   \sum_{i=1}^{n}{-(y_i-mx_i-b)}\\
\end{bmatrix} = \begin{bmatrix}
    \frac{\partial E}{\partial m} \\
    \frac{\partial E}{\partial b} 
\end{bmatrix}
$$
证毕。

至此我们知道，通过求解方程 $X^TXh = X^TY$ 来找到最优参数。这个方程十分重要，他有一个名字叫做 **Normal Equation**，也有直观的几何意义。你可以在 [子空间投影](http://open.163.com/movie/2010/11/J/U/M6V0BQC4M_M6V2AJLJU.html) 和 [投影矩阵与最小二乘](http://open.163.com/movie/2010/11/P/U/M6V0BQC4M_M6V2AOJPU.html) 看到更多关于这个方程的内容。

### 3.4 求解 $X^TXh = X^TY$ 

在3.3 中，我们知道线性回归问题等价于求解 $X^TXh = X^TY$ (如果你选择不做3.3，就勇敢的相信吧，哈哈)

```python
# TODO 实现线性回归
'''
参数：X, Y 存储着一一对应的横坐标与纵坐标的两个一维数组
返回：线性回归的系数(如上面所说的 m, b)
'''
def linearRegression2D(X,Y):
    _X = [[x,1]for x in X]
    _X_T = transpose(_X)
    left_param = matxMultiply(_X_T,_X)
    _Y = [[y]for y in Y]
    righ_param = matxMultiply(_X_T,_Y)
    result = gj_Solve(left_param, righ_param, decPts=4, epsilon=1.0e-16)
    return result[0][0],result[1][0]
```

```python
# 请不要修改下面的代码
m2,b2 = linearRegression2D(X,Y)
assert isinstance(m2,float),"m is not a float"
assert isinstance(b2,float),"b is not a float"
print(m2,b2)
```
